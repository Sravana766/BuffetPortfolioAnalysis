```python
from bs4 import BeautifulSoup
import urllib3
urllib3.disable_warnings()
import pandas as pd
import numpy as np
```

```python
http = urllib3.PoolManager()
r = http.request('GET', 'https://www.dataroma.com/m/holdings.php?m=BRK')
soup = BeautifulSoup(r.data, 'html5lib')
tables = soup.find('table')
```

```python
table_rows=tables.find_all('tr')
table_rows
```

```python
for row in table_rows:
    cols = row.find_all('td')
cols
```

```python
df = pd.DataFrame(columns=["Stock", "Price", "Percentage"])

for row in table_rows:
    cols = row.find_all('td')
    if(cols[1].text.strip() != "Stock"):
        stocks = cols[1].text.strip()
        price = cols[5].text.strip()
        new_price = price.replace("$", "")
        percentage = cols[2].text.strip()
        df = df.append({'Stock':stocks, 'Price':new_price, 'Percentage':percentage}, ignore_index = True) 
df
```

```python
df["Price"] = pd.to_numeric(df["Price"], downcast="float")
df["Percentage"] = pd.to_numeric(df["Percentage"], downcast="float")
df_new = df[df['Price']<150]
df_new
```

```python
df_new["Percentage"] = .100*df_new["Percentage"]
df_new["Percentage"] = .10*df_new["Percentage"]
df_new["Amount"] = 9000*df_new["Percentage"]
df_new
```

```python
df_new['Shares'] = df_new['Amount']/df_new['Price']
df_new
```

```python
df_new['Shares'] = df_new['Shares'].apply(np.floor)
df_new
```

```python
df_final = df_new[df_new['Shares']>=1.0]
df_final
```

```python
amount_being_invested = df_final['Amount'].sum()
amount_remaining = 9000-amount_being_invested
amount_remaining
```

```python
df_final.reset_index(drop=True, inplace=True)
```
